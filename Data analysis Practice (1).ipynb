```python
corr_df=df_tracks.drop(["mode", "explicit", "key"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sn.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black" )
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sn.regplot(data = sample_df, y="loudness", x="energy", color = "c").set(title = "Loundess vs Energy ")
```

```python
plt.figure(figsize=(10,6))
sn.regplot(data = sample_df, y="popularity", x="acousticness", color = "c").set(title = "Popularity vs Acousticness ")
```
